```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_points(shape):
    return np.random.rand(*shape)
```

```python
data1 = (make_points((10000, 2)) - 0.5) * 10
```

```python
def neuronka(data1):
    arr1 = np.sign(data1 @ np.array([-1, 1]) - 1) * (-0.5) + 0.5
    arr2 = np.sign(data1 @ np.array([1, 1]) - 1) * (-0.5) + 0.5

    arr4 = np.sign(
        (np.hstack([arr1.reshape(-1, 1), arr2.reshape(-1, 1)]) @ np.array([1, 1])) - 1.5
    ) * (0.5) + 0.5

    return arr4
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(data1[:,0], data1[:,1], c=neuronka(data1))
```
